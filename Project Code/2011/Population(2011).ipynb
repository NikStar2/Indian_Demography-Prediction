```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

Importing data of 2011


```python
df=pd.read_csv("/content/census_2011_sw.csv")
df
 
```

Re-sizing the plot ratio

```python

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 25
fig_size[1] = 15
plt.rcParams["figure.figsize"] = fig_size
```

Plotting the Population of Male as line plot's 

```python
plt.style.use("ggplot")
plt.plot(df.State[0:10],df.Male[0:10],marker = 's')
plt.xlabel("State")
plt.ylabel("Population of Male in each state")
plt.savefig("img1.jpg")

```
