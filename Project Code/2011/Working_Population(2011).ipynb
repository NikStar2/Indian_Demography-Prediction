```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/content/census_2011_sw.csv")
df
```

```python

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.bar(df.State[0:10],df.Total_Workers[0:10])
plt.xlabel("State")
plt.ylabel("Working Population in each state")
plt.savefig("img1.jpg")
plt.xticks(rotation=45)
```
