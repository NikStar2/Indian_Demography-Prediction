```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/content/census_2011_sw.csv")
df
```

```python

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```
